# üì¶ DATA LOAD

**Data Load** refers to the process of transferring or importing data from one system, source, or file into another system, typically into a database, data warehouse, or analytics platform.

---

## üîç In More Detail:

It is a critical step in **ETL (Extract, Transform, Load)** or **ELT (Extract, Load, Transform)** processes.

---

## üîÑ Types of Data Load:

- **Initial Load** ‚Äì Loading data for the first time.

- **Incremental Load** ‚Äì Loading only new or updated data since the last load.

- **Full Load** ‚Äì Completely replacing the existing data with a new dataset.

## üî∏ Handling Outliers
- Detecting outliers using statistical methods (IQR, Z-score)
- Visualizing with boxplots or scatter plots
- Treating outliers (removal, capping, or transformation)

```python
# Select only numeric columns
df_numeric = df.select_dtypes(include='number')
```

```python
# Now compute IQR-based filtering on numeric data
Q1 = df_numeric.quantile(0.25)
Q3 = df_numeric.quantile(0.75)
IQR = Q3 - Q1
```

```python
# Print IQR information
print("\nQ1:")
print(Q1)
print("\nQ3:")
print(Q3)
print("\nIQR:")
print(IQR)
```

```python
# Filter out the outliers
df_no_outliers = df[~((df_numeric < (Q1 - 1.5 * IQR)) | (df_numeric > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
print("\nDataFrame after removing outliers:")
print(df_no_outliers)
```

## üî∏  Train-Test Split & Model Training (Random Forest)
- Split the dataset into training and testing sets
- Use Random Forest Regressor for predicting Life Expectancy
- Fit the model on training data

```python
# 1. Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.impute import SimpleImputer

# 2. Load dataset
# df = pd.read_csv("Life Expectancy Data.csv")

# 3. Clean column names
df.columns = df.columns.str.strip().str.replace(' +', ' ', regex=True)

# 4. Rename columns if needed
for col in df.columns:
    if 'thinness' in col and '1-19' in col and col != 'thinness 1-19 years':
        df.rename(columns={col: 'thinness 1-19 years'}, inplace=True)

# 5. Encode 'Status' if it exists and not fully null
if 'Status' in df.columns and df['Status'].notna().sum() > 0:
    df['Status'] = df['Status'].map({'Developed': 1, 'Developing': 0})
else:
    print("‚ö†Ô∏è 'Status' column missing or fully null. It will be removed from features.")

# 6. Define features
features = [
    'Adult Mortality', 'infant deaths', 'Alcohol', 'percentage expenditure',
    'Hepatitis B', 'Measles', 'Polio', 'Total expenditure',
    'Diphtheria', 'HIV/AIDS', 'BMI', 'under-five deaths',
    'thinness 1-19 years', 'thinness 5-9 years',
    'Income composition of resources', 'Schooling', 'Status'
]

# 7. Remove features that don't exist or are fully null
features = [col for col in features if col in df.columns and df[col].notna().sum() > 0]
print("‚úÖ Features used:", features)

# 8. Drop rows with missing target
df = df.dropna(subset=['Life expectancy'])

# 9. Prepare X and y
X = df[features]
y = df['Life expectancy']

# 10. Impute missing values in X
imputer = SimpleImputer(strategy='mean')
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

# 11. Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_imputed, y, test_size=0.2, random_state=42
)

# 12. Train model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# 13. Predict and evaluate
y_pred = model.predict(X_test)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

print("\n‚úÖ Model Trained Successfully")
print(f"üìâ RMSE: {rmse:.2f}")
print(f"üìä R¬≤ Score: {r2:.2f}")

# 14. Feature importance
plt.figure(figsize=(10, 6))
pd.Series(model.feature_importances_, index=X.columns).sort_values().plot(kind='barh', color='skyblue')
plt.title("Feature Importance")
plt.xlabel("Importance Score")
plt.tight_layout()
plt.show()
```

## üî∏  Model Evaluation
- Evaluate model performance using metrics like:
  - R¬≤ Score
  - Mean Absolute Error (MAE)
  - Root Mean Squared Error (RMSE)
- Compare predicted vs actual values
- Assess model generalization on test data

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Evaluating model performance
rmse = mean_squared_error(y_test, y_pred, squared=False)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Displaying results
print(f"Root Mean Square Error (RMSE): {rmse:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"R2 Score: {r2:.2f}")
```
